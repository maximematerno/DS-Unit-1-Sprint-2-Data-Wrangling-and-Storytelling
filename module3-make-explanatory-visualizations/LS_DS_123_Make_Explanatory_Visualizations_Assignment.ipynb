```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/inconvenient-sequel/ratings.csv')
print(df.shape)
df.head()
```

```python
df.category.value_counts()
```

```python
df['timestamp']= pd.to_datetime(df['timestamp'])
df.timestamp.describe()
```

```python
df_imdb = df[df.category == 'IMDb users']
df_imdb.shape
```

```python
final = df_imdb.tail(1)
final
```

```python
columns = ['{}_pct'.format(i) for i in range(1,11)]
```

```python
data = final[columns]
data = data.T
data.index = range(1,11)
data
```

```python
plt.style.use('fivethirtyeight')

fig = plt.figure();
fig.patch.set_facecolor('blue');


ax =data.plot.bar(color='C1', width=0.9, legend=False)
ax.set(facecolor='white')

ax.text(x=-1.5, y=50, fontsize=16, fontweight='bold',
         s="'An Inconvenient Sequel: Truth to Power' is divisible")

ax.text(x=-1.3, y=47, fontsize=12,
         s='IMDb ratings for the film as of Aug 29')

ax.set_xticklabels(range(1,11, 1), rotation=0);


ax.text(x=-1.3, y=47, fontsize=12,
         s='IMDb ratings for the film as of Aug 29')


plt.xlabel('Rating', fontsize=10, fontweight='bold')
plt.ylabel('Percent of Total Votes', fontsize=10, fontweight='bold');

```
